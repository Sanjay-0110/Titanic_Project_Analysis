```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\sanja\\Downloads\\train.csv")
```

```python
df.head()
```

```python
#To check Nummber of Rows and Columns

numrows, numcol =df.shape
print("no.of.rows:",numrows)
print("no.of.columns:",numcol)
```

```python
# check for the missing value
Missing_values=df.isnull().sum()
print("Missing value:")
print(Missing_values)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import OneHotEncoder
```

```python
imputer=SimpleImputer(strategy="mean")
df['Age']=imputer.fit_transform(df[["Age"]])

cabin_imputer=SimpleImputer(strategy="most_frequent")
df['Cabin']=cabin_imputer.fit_transform(df[["Cabin"]]).ravel()

embarked_imputer=SimpleImputer(strategy="most_frequent")
df['Embarked']=embarked_imputer.fit_transform(df[["Embarked"]]).ravel()

Missing_values=df.isnull().sum()
print("Missing value:")
print(Missing_values)
```

```python
#Passenger Distribution on ship
passenger_class=df['Pclass'].value_counts()
print("distribution")
print(passenger_class)
```

```python
#embarked count of passenger from each port
embraked=df['Embarked'].value_counts()
print("people onboard for each port:")
print(embraked)
```

```python
Survived_count=df['Survived'].sum()
print("no.of.people survived:",Survived_count)
```

```python
#survival rate
survival_rate=df['Survived'].mean()*100
print('survival rate:{:2f}%'.format(survival_rate))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='Sex',y='Survived',data=df)
plt.title("Gender based survived")
plt.savefig('Gender based survived.png')
plt.show()
```

```python
sns.barplot(x='Pclass',y='Survived',data=df)
plt.title("class wise survived")
plt.savefig('class survived graph')
plt.show()
```

```python
sns.barplot(x='Embarked',y='Survived',data=df)
plt.title("Embarked survived rate")
plt.savefig("embarked survived.png")
plt.show()
```

```python
#age distribution
sns.histplot(df['Age'],bins=20,kde=True)
plt.title("age distribution")
plt.xlabel("age")
plt.ylabel("count")
plt.savefig("age distribution")
plt.show()
```

```python
# age wise survived people graph
df['Agegroup']=pd.cut(df['Age'],bins=[0,18,30,50,df['Age'].max()],labels=['0-17','18-29','30-49','50+'])

#graph
sns.barplot(x='Agegroup',y='Survived',data=df)
plt.title("age survived")
plt.savefig("age survived")
plt.show()
```

```python
df['FamilySize']=df['SibSp']+df['Parch']+1

#plotting the graph
sns.barplot(x='FamilySize',y='Survived',data=df)
plt.title('familysize vs survived')
plt.savefig("familysize")
plt.show()
```
